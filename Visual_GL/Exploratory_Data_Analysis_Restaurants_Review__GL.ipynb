```python
import csv

# Open and read the positive file
with open('positive.txt', 'r', encoding='utf-8', errors='backslashreplace') as f:
    reader = csv.reader(f)
    positive_words = [word[0] for word in reader]

# Open and read the negative file
with open('negative.txt', 'r', encoding='utf-8', errors='backslashreplace') as f:
    reader = csv.reader(f)
    negative_words = [word[0] for word in reader]
```

```python
## get_dataset(category)_get dataset by category
def get_dataset(category):
    df = DA_Reviews[['removed_punct_text', 'labels']][DA_Reviews.category == category]
    df.reset_index(drop=True, inplace=True)
    df.rename(columns={'removed_punct_text': 'text'}, inplace=True)
    return df


## filter_words (reviews)_import negative file which contains common meaningless positive words such as bad
def filter_words(review):
    words = [word for word in review.split() if word in positive_words + negative_words]
    words = ' '.join(words)
    return words
```

```python
## filter words
def filter_words(review, positive_words, negative_words):
    # Tokenize the review into words
    words = review.split()
    # Filter out words that are not in positive_words or negative_words
    filtered_words = [word for word in words if word in positive_words or  word in negative_words]
    # Join the filtered words back into a single string
    filtered_review = ' '.join(filtered_words)
    return filtered_review

```

```python

American_train['text'] = American_train['text'].apply(filter_words, positive_words=positive_words, negative_words=negative_words)

print(American_train['text'])
```

# Analyze polarity score of each word in the specific category

American

```python
# Get the coefficients (weights) of the features (words) from the SVM model
coeff = svm.coef_[0]

# Create a DataFrame 'American_words_score' containing word scores and their corresponding words
American_words_score = pd.DataFrame({'score': coeff, 'word': vectorizer.get_feature_names_out()})
```

```python
## get frequency of each word in all reviews in specific category
American_reviews = pd.DataFrame(feature_train_counts, columns=vectorizer.get_feature_names_out())
American_reviews['labels'] = class_train
American_frequency = American_reviews[American_reviews['labels'] =='positive'].sum()[:-1]
```

```python
# Set the 'word' column as the index of the DataFrame 'American_words_score'
American_words_score.set_index('word', inplace=True)
```

```python
American_polarity_score = American_words_score
American_polarity_score['frequency'] = American_frequency
```

```python
# Calculate the polarity score for each word in the 'American' category
American_polarity_score['polarity'] = American_polarity_score['score'] *American_polarity_score['frequency'] / American_reviews.shape[0]

print(American_polarity_score['score'] *American_polarity_score['frequency'])
print(American_reviews.shape)
```

```python
# Select rows (words) from 'American_polarity_score' DataFrame based on the lists of unuseful words
unuseful_positive_words = American_polarity_score.loc[['great', 'amazing', 'love', 'best', 'awesome', 'excellent', 'good',
                                                     'favorite', 'loved', 'perfect', 'gem', 'perfectly', 'wonderful',
                                                     'happy', 'enjoyed', 'nice', 'well', 'super', 'like', 'better', 'decent', 'fine',
                                                     'pretty', 'enough', 'excited', 'impressed', 'ready', 'fantastic', 'glad', 'right',
                                                     'fabulous']]

unuseful_negative_words =  American_polarity_score.loc[['bad', 'disappointed', 'unfortunately', 'disappointing', 'horrible',
                                                      'lacking', 'terrible', 'sorry', 'disappoint']]

# Drop the rows (words) from 'American_polarity_score' DataFrame corresponding to unuseful words
American_polarity_score.drop(unuseful_positive_words.index, axis=0, inplace=True)
American_polarity_score.drop(unuseful_negative_words.index, axis=0, inplace=True)
```

```python
# Convert the 'polarity' column to float data type
American_polarity_score['polarity'] = American_polarity_score['polarity'].astype(float)

# Convert the 'frequency' column to float data type
American_polarity_score['frequency'] = American_polarity_score['frequency'].astype(float)
```
