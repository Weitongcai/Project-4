```python
import pandas as pd
```

```python
#read in CSA dataset
df = pd.read_csv('yelp_dataset.csv')
df.head()
```

```python
#find the top 1 category
df.dropna(subset=['category'], inplace=True)

categories_column = df['category'].str.split(',')

flattened_categories = [category.strip() for categories_list in categories_column for category in categories_list]
category_counts = pd.Series(flattened_categories).value_counts()
top_category = category_counts.index[0]

print("Top 1 Category:", top_category)
```

```python
#read in review dataset
df_reviews = pd.read_csv('yelp_reviews.csv')
df_reviews
```

```python
#find the top 1 category again to testify previous top1 category
df_reviews.dropna(subset=['category'], inplace=True)

categories_column = df_reviews['category'].str.split(',')

flattened_categories = [category.strip() for categories_list in categories_column for category in categories_list]
category_counts = pd.Series(flattened_categories).value_counts()
top_category = category_counts.index[0]

print("Top 1 Category:", top_category)
```

```python
#select only columns needed and drop rows with missing data
selected_columns = ['business_id', 'latitude', 'longitude', 'stars', 'review_count', 'category', 'review_id', 'user_id']
df_selected = df_reviews[selected_columns]
df_selected.dropna(inplace=True)
df_selected.drop(df_selected[df_selected['review_count'] < 10].index, inplace=True)
df_selected
```

```python
#choose users that have at least 5 reviews
user_review_counts = df_selected.groupby('user_id')['review_id'].count()
filtered_user_ids = user_review_counts[user_review_counts > 5].index
df_filtered = df_selected[df_selected['user_id'].isin(filtered_user_ids)]
df_filtered
```

```python
df_filtered.to_csv('cleaned_data.csv', index=False)
```
