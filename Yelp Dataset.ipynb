```python
import json
import pandas as pd
```

#  Clean yelp_business dataset

```python
data = []  # Create an empty list to store the parsed JSON data

# Read JSON data from a file and append each line (parsed as JSON) to the 'data' list
with open('yelp_academic_dataset_business.json', encoding='utf-8') as json_file:
    for line in json_file:
        data.append(json.loads(line))

# Read JSON data from a file or JSON string
yelp_df = pd.DataFrame(data)
# 'data' is now a DataFrame containing the JSON data
yelp_df.tail()
```

```python
#drop the columns we don't need for this analysis
yelp_df = yelp_df.drop(columns=["is_open", "attributes"])
yelp_df.head()
```

```python
## filter restaurants of 50 states from USA
states = ["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DC", "DE", "FL", "GA", 
          "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD", 
          "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ", 
          "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC", 
          "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"]
yelp_usa=yelp_df.loc[yelp_df['state'].isin(states)]
yelp_usa.tail()
```

```python
## drop null values in category, delete original column categories and reset the index
yelp_usa = yelp_usa.dropna(subset=['categories']).copy()
#del yelp_usa['categories']
usa_restaurants = yelp_usa.reset_index(drop=True)
usa_restaurants.head(10)

# Select all rows where the 'categories' column contains the substring 'Restaurants'
usa_restaurants = usa_restaurants[usa_restaurants['categories'].str.contains('Restaurants')]
usa_restaurants.tail()
```

```python
# label reviews 
usa_restaurants['labels'] = ''
usa_restaurants.loc[usa_restaurants.stars >=4, 'labels'] = 'Excellent'
usa_restaurants.loc[usa_restaurants.stars ==3, 'labels'] = 'Great'
usa_restaurants.loc[usa_restaurants.stars <3, 'labels'] = 'Ok'

usa_restaurants.head()
```
